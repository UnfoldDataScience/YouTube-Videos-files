# Encoder Only Model

```python
from transformers import BertTokenizer, BertModel
import torch
```

```python
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
```

```python
inputs = tokenizer("Transformers are powerful!", return_tensors="pt")
```

```python
model = BertModel.from_pretrained("bert-base-uncased")
```

```python
outputs = model(**inputs)
```

```python
print(outputs.last_hidden_state.shape)
```

```python
outputs.last_hidden_state
```

#Decoder only Model

```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel
```

```python
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
```

```python
model = GPT2LMHeadModel.from_pretrained("gpt2")
```

```python
input_ids = tokenizer("Once upon a time,", return_tensors="pt").input_ids
```

```python
output = model.generate(input_ids, max_length=20, do_sample=True)
```

```python
output
```

```python
print(tokenizer.decode(output[0]))
```

# Encode Decoder Model

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration
```

```python
model = T5ForConditionalGeneration.from_pretrained("t5-small")
```

```python
tokenizer = T5Tokenizer.from_pretrained("t5-small")
```

```python
input_text = """summarize: In today's fast-changing world, students must adopt a mindset of continuous learning.
New technologies, shifting industries, and global economic changes mean that knowledge can become outdated quickly.
Focusing on lifelong learning equips students with the skills to adapt and remain competitive, no matter how the job market evolves."""
```

```python
input_ids = tokenizer(input_text, return_tensors="pt").input_ids
```

```python
output = model.generate(input_ids)
```

```python
print(tokenizer.decode(output[0], skip_special_tokens=True))
```
